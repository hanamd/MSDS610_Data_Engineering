## Unsupervised learing 

Using unsupervised learning to cluster data points together since they don't have labels of low, medium, or high risk.

```python
#!pip install scikit-learn
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report # text report showing the main classification metrics.
from sklearn.cluster import KMeans
```

in the previous lab, we only used the features we selected for our model to perform KMeans clustering. However, this time I wanted to try using all the columns so that KMeans can find a pattern for the clusters based on all the columns.

```python

# selected features
features = gb[['Temperature_Anomaly', 'CO2_Emissions', 'Population',
       'Forest_Area', 'GDP', 'Renewable_Energy_Usage', 'Methane_Emissions',
       'Sea_Level_Rise', 'Arctic_Ice_Extent', 'Urbanization',
       'Deforestation_Rate', 'Extreme_Weather_Events', 'Average_Rainfall',
       'Solar_Energy_Potential', 'Waste_Management', 'Per_Capita_Emissions',
       'Industrial_Activity', 'Air_Pollution_Index', 'Biodiversity_Index',
       'Ocean_Acidification', 'Fossil_Fuel_Usage',
       'Energy_Consumption_Per_Capita', 'Policy_Score', 'Average_Temperature']]
```

```python
#  K means custering with 3 clusters ( Low, Medium, High risk)

kmeans = KMeans(n_clusters=3, random_state=42)

gb['Risk_Level'] = kmeans.fit_predict(features)

# 0: 'Low', 1: 'Medium', 2: 'High'
```

## Random Forest Model 1

For the first model, we will use all the columns from our dataset to create the model. Afterwards, we will analyze the important features selected from the model's performance to run another model.

```python
# Random Forest

X = features

y = gb['Risk_Level'] # Dependent


# Split data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Intitalize Rnd

rnd = RandomForestClassifier(n_estimators=100, random_state=42)


# Model Train

rnd.fit(X_train, y_train)
```

```python
# predict

y_pred = rnd.predict(X_test)

print(classification_report(y_test, y_pred))
```

## Analyze Feature importance

```python
# method 1
feat_imp = pd.Series(rnd.feature_importances_, index=X.columns).sort_values(ascending=False)
feat_imp
```

## Random Forest Model 2

From the feature importance list of our first model, we learned that GDP has a big feature importance. Because of that, for our new model, we will use our top 2 features (CO2_Emissions, Methane_Emissions) from our previous lab and also include GDP. <br>
You will also notice this time we also split out dataset int train,test and **Validation** set.

```python
# Random Forest 2

X = gb[['CO2_Emissions', 'Methane_Emissions','GDP']]

y = gb['Risk_Level'] # Dependent


# Split data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)


X_val, X_hold, y_val, y_hold = train_test_split(X_test, y_test, test_size=0.5, random_state=42, stratify=y_test)


# Print dataset sizes
print("Training set:", X_train.shape)
print("Validation set:", X_val.shape)
print("Test set:", X_test.shape)

# Intitalize Rnd

rnd2 = RandomForestClassifier(n_estimators=100, random_state=42)


# Model Train

rnd2.fit(X_train, y_train)
```

```python
# predict

y_pred = rnd2.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# method 1
feat_imp = pd.Series(rnd2.feature_importances_, index=X.columns).sort_values(ascending=False)
feat_imp
```
